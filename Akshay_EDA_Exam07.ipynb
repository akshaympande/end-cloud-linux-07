<h1>Ratio = Total Income / Anumity amount </h1>

```python
merge_df['INCOME_ANNUITY_RATIO'] = merge_df['AMT_INCOME_TOTAL'] / merge_df['AMT_ANNUITY_x']
suspicious_profiles_income_ratio = merge_df[merge_df['INCOME_ANNUITY_RATIO'] < 1]  # Below a certain threshold
print(suspicious_profiles_income_ratio)
```

```python
len(suspicious_profiles_income_ratio)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for INCOME_ANNUITY_RATIO by TARGET
plt.figure(figsize=(10, 6))
sns.boxplot(data=merge_df[merge_df['INCOME_ANNUITY_RATIO'] < 1], x='TARGET', y='INCOME_ANNUITY_RATIO', palette=['green', 'red'])
plt.title('Income-to-Annuity Ratio by Target for Suspicious Profiles', fontsize=16)
plt.xlabel('TARGET (0: Non-Fraud, 1: Fraud)', fontsize=14)
plt.ylabel('Income-to-Annuity Ratio', fontsize=14)
plt.show()


# TARGET (0: Non-Fraud, 1: Fraud)

#There are 141 People having lesser ratio are the suspeciuos profile Can have high chance of performing fraud 
```
